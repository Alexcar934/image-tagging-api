```python
import requests
import base64

# 1. Carga y codifica en base64 la imagen (asegúrate de que está en tu ruta local)
with open("cerveza_small.jpg", "rb") as f:
    img_b64 = base64.b64encode(f.read()).decode("utf-8")

# 2. Construye el JSON con la imagen
payload = {"data": img_b64}

# 3. Haz la petición POST al endpoint (ajusta el puerto si es diferente)
response = requests.post(
    "http://localhost:8080/image?min_confidence=80",
    json=payload
)

# 4. Imprime los resultados
print("Status code:", response.status_code)
try:
    print("Response JSON:", response.json())
except Exception:
    print("Error en respuesta:", response.text)
```

```python
import requests
# 1. Carga y codifica en base64 la imagen (asegúrate de que está en tu ruta local)
with open("cerveza.jpg", "rb") as f:
    img_b64 = base64.b64encode(f.read()).decode("utf-8")

# 2. Construye el JSON con la imagen
payload = {"data": img_b64}

# 3. Haz la petición POST al endpoint (ajusta el puerto si es diferente)
response = requests.post(
    "http://localhost:8080/image?min_confidence=80",
    json=payload
)

# 4. Imprime los resultados
print("Status code:", response.status_code)
try:
    print("Response JSON:", response.json())
except Exception:
    print("Error en respuesta:", response.text)

    
image_id = response.json()['id'] # sustituye con el UUID real
resp = requests.get(f"http://localhost:8080/image/{image_id}")

print("Status:", resp.status_code)
print("Response:", resp.json())
```
